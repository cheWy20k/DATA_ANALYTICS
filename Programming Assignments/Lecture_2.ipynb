```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ds = 'lymphography.data'
column_names = ["class", "lymphatics", "block_of_affere", "bl.aff.asc", "bl.aff.desc", "popliteal", "int.iliac", "femoral"]
df = pd.read_csv(ds, names=column_names)

plt.figure(figsize=(8, 6))
df['class'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='block_of_affere', y='popliteal', hue='class')
plt.title('Block of Affere vs Popliteal')
plt.xlabel('Block of Affere')
plt.ylabel('Popliteal')
plt.legend(title='Class')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='class', y='block_of_affere', estimator='mean', err_style=None)
plt.title('Block of Affere Across Classes')
plt.xlabel('Class')
plt.ylabel('Mean Block of Affere')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='block_of_affere')
plt.title('Comparison: Block of Affere Across Classes')
plt.xlabel('Class')
plt.ylabel('Block of Affere')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Plot')
plt.show()
```
